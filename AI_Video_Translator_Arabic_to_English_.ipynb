```python
!pip install faster-whisper
```

```python
from faster_whisper import WhisperModel

model = WhisperModel(
    "turbo",
    device="cpu",
    compute_type="int8"
)

segments, info = model.transcribe(
    "output.wav",
    language="ar",
    vad_filter=True,
    initial_prompt="""
    نظام ERP، موديول المبيعات، Workflow،
    Dashboard، الفاتورة، حركة المبيعات
    """
)

full_text = ""

for segment in segments:
    full_text += segment.text + " "

print(full_text)
```

```python
arabic_text = full_text

print(arabic_text)
```

```python
import os
import requests

API_KEY = os.getenv("OPENROUTER_API_KEY")

url = "https://openrouter.ai/api/v1/chat/completions"

headers = {
    "Authorization": f"Bearer {API_KEY}",
    "Content-Type": "application/json"
}

data = {
    "model": "~openai/gpt-mini-latest",
    "max_tokens": 800,
    "messages": [
        {
            "role": "user",
            "content": f"""
Fix this Arabic ASR text and translate it into natural professional English for voice dubbing:

{arabic_text}
"""
        }
    ]
}

response = requests.post(url, headers=headers, json=data)

print("STATUS:", response.status_code)
print("RESPONSE:", response.text)
```

```python
response = requests.post(url, headers=headers, json=data)

# تحويل الرد لـ JSON
result = response.json()

# استخراج النص
full_text = result["choices"][0]["message"]["content"]

print(full_text)
```

```python
!pip install edge-tts
```

```python
import edge_tts

english_text = {full_text}


async def generate_voice():
    communicate = edge_tts.Communicate(
        text=english_text,
        voice="en-US-JennyNeural"
    )

    await communicate.save("voice.mp3")

# مهم في Colab
await generate_voice()

print("Natural voice generated!")
```

```python
from pydub import AudioSegment

# تحميل الملفات
music = AudioSegment.from_file("music.mp3")
voice = AudioSegment.from_file("voice.mp3")

# رفع صوت الموسيقى شوي
music = music - 5

# إضافة صمت 3 ثواني قبل الكلام
silence = AudioSegment.silent(duration=3000)

# تأخير الصوت
voice_delayed = silence + voice

# دمج الموسيقى مع الصوت
final_audio = music.overlay(voice_delayed)

# حفظ النتيجة
final_audio.export("final_audio.mp3", format="mp3")

print("Audio mixed successfully!")
```

final video

```python
from moviepy.editor import VideoFileClip, AudioFileClip

# تحميل الفيديو الأصلي
video = VideoFileClip("input.mp4")

# تحميل الصوت النهائي
audio = AudioFileClip("final_audio.mp3")

# تركيب الصوت على الفيديو
final_video = video.set_audio(audio)

# حفظ الفيديو النهائي
final_video.write_videofile(
    "final_output.mp4",
    codec="libx264",
    audio_codec="aac"
)

print("Final dubbed video created!")
```
